# Linear Regression
 
 $ Y_i = \beta_0 + \beta_1  X_i + \varepsilon_i $
 
  __Car Weight & Mileage__



Although it is relatively simple compared to many modern data analysis techniques, linear regression is an incredibly useful tool in statistics. 



Let's walk through a quick example together

```python
#import basic libraries
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

Next we'll load the data which can be found here in the attached file: __link__

```python
#load data
car_data = pd.read_csv('MPGData.txt',delimiter=' ')
car_data.head(10)
```

Looks like we read in the data successfully!

Let's explore the data a little bit with a simple scatter plot. If the data appear to have a linear relationship, then we should be able to proceed using simple linear regression.

```python
#visualize data


%config InlineBackend.figure_format = 'retina'
X =  car_data.drop('MPG',axis = 1) #we'll use the weight of the cars as our X variable
Y =  car_data.MPG #we'll use the mileage of the car as our Y variable
plt.scatter(X,Y)
plt.xlabel('Car Weight')
plt.ylabel('Mileage')
plt.show()
```

Looking at this plot, I think it's pretty safe to say that the weight and mileage of a car are indeed linearly related. So we'll go ahead and proceed with our regression analysis.

When doing simple linear regression, there are four basic assumptions or rules that we need to follow. We can easily remember them by using the acronym "LINE". 

__L - Linear:__ The data needs to be linearly related.
- It wouldn't make sense for us to try and model data with a straight line if our data isn't linearly related.
- We can verify that our data is linearly related with our scatterplot above.
- We can also assess the strength of that linearity by calculating the correlation coefficient.

__I - Independent:__ The error terms $ \varepsilon_1, \varepsilon_2, \ldots, \varepsilon_n$ are independant random variables.
- For two variables to be independent means that having information about one variable doesn't give you any information about the other.
- For more information on independence see wikipedia.
- I haven't found a good graphic or visual aid to test for independence, but for  the most part we can safely make this assumption by taking a step back and critically thinking about the data
- For example: If I know that you wore a blue shirt today, that doesn't tell me anything about the liklihood of you getting in a traffic accident on your way home from work. So your shirt color and the traffic conditions would be considered independant. However, let's say that I learned that your hometown was hit by a blizzard today. That would lead me to know that there is a greater chance of you getting in a traffic accident on your way home. Therefore weather and traffic conditions are not independent of eachother.

__N - Normal:__ The error terms follow a normal distribution.
- We can check this assumption using a histogram, a Q-Q Plot or a KS test.

__E - Equal Variance__
- This means that the variance of the error terms must be constant throughout the entire data set.
- The big fancy word for this is "homoscedastic"
- We can check this assumption using the residual plots

With that, let's fit a model to the data set and see if we meet our assumptions! We will use sklearn to fit our model. 

```python
#using sklearn
from sklearn import linear_model as lm

#create a linear regression object (regr is short for regressor)
regr = lm.LinearRegression()
regr
```

```python
#fit the model to the data
regr.fit(X,Y)
```

```python
#model summary
print(' Intercept: %f\n'% regr.intercept_,'Coefficient: %f' % regr.coef_[0])
```

Our regression model is set up as follows: 

 $ Y_i = \beta_0 + \beta_1  X_i + \varepsilon_i $


$Y_i$ is the mileage of the **ith** car.

$X_i$ is the weight of the **ith** car.



The intercept, $\beta_0$, is the expected mileage of a car that weighs 0lbs. In this situation, the interpretation isn't particularly useful, but it's still important to understand what it means. 

The coefficient, $\beta_1$, tells us the average change in mileage for every increase of 1lb in car weight. 

```python
regr.score(X,Y)
X.min()[0]
```

```python
residuals = Y - regr.predict(X)
plt.scatter(X,residuals)
plt.hlines(y=0,xmin = X.min()[0],xmax=X.max()[0],color = 'red')
plt.show()
plt.hist(residuals)
plt.show()
```

```python
#Residual Plots
sns.residplot(X,Y)
sns.despine() #removes excess info from the plot
```

```python
#plot of the model fitted to the data
plt.scatter(X,Y)
plt.plot(X, regr.predict(X),color = 'red',linewidth = 3)
plt.show()
```

 Our model for predicting is as follows: 
 
 
 $ \hat{y} = \beta_0 + \beta_1  X_i $

```python
#semi truck: 80,000 lbs
regr.predict(6000)
```

```python
#RPMSE
regr.score(X,Y)
```
